```python
# import libraries
from urllib.request import urlopen
from bs4 import BeautifulSoup
import csv
```

```python
url_list = []
```

```python
for i in range(0, 11):
    url_list.append("http://www.police.ufl.edu/wp-content/asp/crimelog/search.asp?intpage=" + str(i) + "&action=%3E&sortBy=&searchType=Type%20of%20Crime&searchField=sexual%20battery")
```

```python
data_list = []
```

```python
for i in range(0, len(url_list)):
    page = urlopen(url_list[i])
    soup = BeautifulSoup(page, "html.parser") # parse the html using beautiful soup and store in variable `soup`
    divs = soup.findAll("table", {"border": "1"}) # Take out the <div> of otherInfo and get its value
    x = len(data_list)
    for div in divs:
        rows = div.findAll('tr')
        for row in rows:
            data_list.append(row.findAll('td'))
    for j in range(x, len(data_list)):
        for k in range(0, len(data_list[j])):
            data_list[j][k] = data_list[j][k].text.strip()
```

```python
data_list
```

```python
# open a csv file with append, so old data will not be erased
with open('batteryLocations.csv', 'a') as f:
    writer = csv.writer(f)
    writer.writerow(["Date", "Location", "Disposition"]) # write csv file headers
    for i in range (0, len(data_list)):
        writer.writerow(data_list[i][-3:])
```
